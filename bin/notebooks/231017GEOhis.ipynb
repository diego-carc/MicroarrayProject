```python
import pandas as pd
import matplotlib.pyplot as plt
geo = pd.read_csv("../results/231018GEOmetadata/esummary_metadata_2.tsv", sep="\t").set_index("Unnamed: 0")
```

```python
gpls = {gsm for gsms in geo.GPL.to_list() for gsm in str(gsms).split(';')}
```

```python
from re import sub
from os import remove
from GEOparse import get_GEO
from functions import get_GEO_meatadata
def get_geo_object(ID, path, tries = 10):
  try:
    geo = get_GEO(geo=ID, destdir=path, silent=True)
    remove(f'{path}/{ID}.txt')
    return(geo)
  except:
    if not tries: return(None)
    print(ID, tries)
    return(get_geo_object(ID, path, tries-1))
  
def frmt(num_id):
    num_id = str(int(num_id))
    if num_id.startswith('1'): gtyp = "GPL" ; i='1'
    elif num_id.startswith('2'): gtyp = "GSE"; i ='2'
    else: gtyp = "GSM"; i = '3'
    return(sub(f'^{i}0+', gtyp, num_id))
```

```python
gpls_10 = [frmt(gpl) for i,gpl in enumerate(gpls) if i < 10]
```

```python
gpls_10
```

```python
a = {gpl : get_geo_object(gpl,'\.', 15) for gpl in gpls_10}
```

```python
b = {k : {i: ';'.join(j) for i,j in v.metadata.items()} if v else {} for k,v in a.items()}
```

```python
pd.DataFrame.from_dict(b).T.reset_index().rename(columns={'index':'GPL'}).to_csv("./caballodefuego.tsv", sep='\t')
```
